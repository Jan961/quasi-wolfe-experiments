```python
import numpy as np
import matplotlib.pyplot as plt
from function_generation.PolynomialGenerator import PolynomialGenerator
```

```python
# Generate a polynomial function
f,g,h = PolynomialGenerator().generate(2,3)

margin =0.5
samples_per_side = 100

x = np.linspace(-1-margin,1+margin,samples_per_side)
y = np.linspace(-1-margin,1+margin,samples_per_side)


# plot the function

mesh = np.meshgrid(x, y)


ys = f(mesh)
plt.imshow(ys, extent=(-1-margin,1+margin,-1-margin,1+margin), origin='lower')

```
