```python
import numpy as np
import matplotlib.pyplot as plt

# Define the points and derivatives
x1, y1, y1_prime = 1, 2, 3
x2, y2, y2_prime = 4, 5, 6

# Construct the coefficient matrix
A = np.array([
    [x1**3, x1**2, x1, 1],
    [x2**3, x2**2, x2, 1],
    [3*x1**2, 2*x1, 1, 0],
    [3*x2**2, 2*x2, 1, 0]
])

# Construct the right-hand side vector
b = np.array([y1, y2, y1_prime, y2_prime])

# Solve the system of equations for the coefficients
coefficients = np.linalg.solve(A, b)

# The coefficients are a, b, c, d
a, b, c, d = coefficients

print(f"The cubic polynomial is: f(x) = {a}x^3 + {b}x^2 + {c}x + {d}")

# Define the cubic function
def cubic_function(x):
    return a*x**3 + b*x**2 + c*x + d

# Generate x values for plotting
x_values = np.linspace(min(x1, x2) - 1, max(x1, x2) + 1, 400)
y_values = cubic_function(x_values)

# Plot the function
plt.plot(x_values, y_values, label=f'f(x) = {a:.2f}x³ + {b:.2f}x² + {c:.2f}x + {d:.2f}')

# Plot the given points
plt.scatter([x1, x2], [y1, y2], color='red', zorder=5)
plt.quiver([x1, x2], [y1, y2], [1, 1], [y1_prime, y2_prime], color='blue', angles='xy', scale_units='xy', scale=1, zorder=5)

# Label the points
plt.text(x1, y1, f'({x1}, {y1})', fontsize=9, verticalalignment='bottom')
plt.text(x2, y2, f'({x2}, {y2})', fontsize=9, verticalalignment='bottom')

# Add title and labels
plt.title('Cubic Function Fit to Two Points with Known Derivatives')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
